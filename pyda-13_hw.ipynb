```python
# Задание 1
# Используйте файл с оценками фильмов ml-latest-small/ratings.csv. 
# Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. 
# Под временем жизни понимается разница между максимальным и минимальным 
# значением столбца timestamp для данного значения userId.



import pandas as pd

ratings = pd.read_csv('ml-latest-small/ratings.csv')

new_ratings = ratings.groupby('userId').agg({'timestamp': ['min', 'max'], 'movieId': 'count'})

new_ratings = new_ratings[ new_ratings['movieId']['count'] > 100 ]

new_ratings['life_time'] = new_ratings['timestamp']['max'] - new_ratings['timestamp']['min']

new_ratings['real_lifetime_months'] = pd.to_timedelta(new_ratings['life_time'], unit='s').map(lambda x: round(x.days/30, 2))

print('Среднее время жизни пользователя, который выставил более 100 оценок (в месяцах): ')
new_ratings.real_lifetime_months.mean()


```

```python
# Задание 2
# Дана статистика услуг перевозок клиентов компании по типам
# Необходимо сформировать две таблицы:
# -таблицу с тремя типами выручки для каждого client_id без указания адреса клиента
# -аналогичную таблицу по типам выручки с указанием адреса клиента

rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
# решение 

table_1 = rzd.merge(auto, how='outer', on='client_id').merge(air, how='outer', on='client_id').fillna(0)
table_1
```

```python
table_2 = table_1.merge(client_base, how='left', on='client_id')
table_2[ ['client_id', 'address', 'rzd_revenue', 'auto_revenue', 'air_revenue'] ]
```
